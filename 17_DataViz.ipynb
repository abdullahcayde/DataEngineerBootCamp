# Final ZOOM Plot

```python
plt.grid(True)

x = np.arange(-5, 5, .01)

# Buyuk grafikdekiler
plt.plot(x, x*x, label='$x^2$')
plt.plot(x, np.exp(-x), 'g--', label='$exp(-x)$')

plt.xlim(-5,2)
plt.ylim(0,25)

# Ok ve yazi , legend
plt.annotate('Zoom', xy=(-1.8, 7), xytext=(-.5,6), arrowprops={'facecolor':'red'})

plt.title('My summary graph')

plt.legend()

# Kirmizi Plot ici Bos
plt.plot([-3, -2, -2, -3, -3],[5, 5, 10, 10, 5],'r', alpha = 0.6)


# Plot Zoom
plt.axes([.55, 0.4, .2, .2])

plt.plot(x,x**2,'b')
plt.plot(x,np.exp(-x),'g--')

plt.axis([-3,-2,5,10])

plt.title("Plot Zoom")
plt.xticks([])
plt.yticks([]);
```

```python
x = np.linspace(-6, 2, 40)
# A sequence of 40 points is created in the interval[-6, 2]

plt.grid(True, linestyle ='--')
# A grid is added to the graph

plt.plot(x, x**2,'b',label='$x^2$')
# We draw the line plot x^2 and give it the corresponding label for the legend.

plt.plot(x,np.exp(-x),'g--', label='$exp(-x)$')
# The exp(-x) line plot is drawn in dotted lines and given the corresponding label for the legend.

plt.axis([-5,2,0,25])
# We modify the amplitude of the axes

plt.plot([-3, -2, -2, -3, -3],[5, 5, 10, 10, 5],'r', alpha = 0.6)
# We create the square that surrounds the part of the graph that we will reproduce
# The'alpha' argument gives the percentage of opacity of the plot (1 opaque, 0 invisible transparent)

plt.annotate('Zoom', xy=(-1.8, 7.5), xytext=(-0.5, 7.5), 
            arrowprops={'facecolor':'red'} )
# The red arrow is created, with the text'Zoom' pointing to the point (-1.8, 7.5).

plt.title('My summary graph')
#The graph is given a title

plt.legend()
# Legends are displayed

plt.axes([.55, 0.4, .2, .2])
# A new graph is created inside the previous one,
# whose lower left corner starts at the point (0.55, 0.4) in relative distance, 
# where 0 represents the origin, and 1 represents the end of the axis.
# This graph will have a width and height representing 20% of the width 
# and the height of the original graph.

#From then on, the'pyplot' commands will only apply to this new graph.

plt.plot(x,x**2,'b')
plt.plot(x,np.exp(-x),'g--')
# We draw the same line plot as before.

plt.axis([-3,-2,5,10])
# But we limit the axes to the coordinates of the square we have drawn.

plt.xticks([])
plt.yticks([]);
# Delete the graduations on the new graph.
```

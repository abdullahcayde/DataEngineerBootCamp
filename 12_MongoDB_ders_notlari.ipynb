# Bolum 1 - NoSQL databases (MongoDB)

```python
NoSQL databases (MongoDB)


1. Introduction to MongoDB
a) What is NoSQL?
Relational databases have been used since the 1970s. They are widely used and well documented and are very useful to represent relationships between different entities. However, they require a strict schema. Nowadays, developers have storage systems that are simpler to use.

With Big Data, we are faced with a multitude of different types of data. For example, if we want to store the activity of a user on a website. This website will evolve, have new pages, new features. If we want to track the use of these new features, we may need to update the schema of our database.

Recently, the focus has been on NoSQL databases which allow for greater flexibility on the data schema and thus manage data with a wide range of formats.

  Originally, NoSQL means No SQL, which means, database management systems that are not based on the relational paradigm. It has been Not Only SQL (not just SQL) as the boundary between the two families has become blurrier.
NoSQL DBMSs are made up of a wide range of technologies, techniques and paradigms, but there are three main families:

Document-oriented databases

Column-oriented databases

Graph-oriented databases

Below, we will detail what a document-oriented database is.

b) Document-oriented databases
In a relational database, data is stored in rows and columns, forming a table, following a strict schema.

{"menu": {
    "sandwichs": [
        {"name": "Melthouse","bread":"Wheat","sauce": "Ceasar"},
        {"name": "Italian BMT", "extras": ["pickles","onions","lettuce"],"sauce":["Chipotle","Aioli"]},
        {"name": "Steakhouse Melt","bread":"Parmesan Oregano"}
}}
It would be complex to store them in a relational database. Indeed, there is no strict schema and we notice that our data (here sandwiches) have different attributes.

However, this is not the case with document-oriented databases where we will store information in documents. A document is a set of key-values. That is why these databases are adequate to process our semi-structured data.

These documents are then organized into what we call collections. In this way, we make sure that similar documents are grouped together. Finally, we put our collections into databases. This way, we can organize our data.

The following tools provide document-oriented databases:

Couchbase

CouchDB

MongoDB

We will see in particular MongoDB which is the reference in this field.

c) Installation
We will use Docker to handle MongoDB in order to avoid wasting time during the installation.

Write the following lines in a docker-compose.yml file:

version: "3.3"
services:
  mongodb:
    image : mongo:5.0
    container_name: my_mongo
    volumes:
    - ./sample_training:/data/db
    ports:
    - 27017:27017

  It is possible to associate a user to our MongoDB server. To do this, we can use this file instead.
Run the following command:

docker-compose up -d 
With the docker ps command, we can verify that our MongoDB container is operational. Now, let's switch to the environment with Mongo using the following command:

docker exec -it my_mongo bash 
We are in a different environment from our personal machine, we can write mongo and press the tab key to see that several commands dedicated to Mongo are available whereas they were not there initially.

root@47359760cda5:/# mongo
mongod        mongodump     mongoexport   mongofiles    mongoimport   mongorestore  mongos        mongosh       mongostat     mongotop 
We will see how to manipulate data on MongoDB in the next chapter. To exit the MongoDB environment, simply issue the exit command.
```

# Bolum 3 - Advanvanced MongoDB Queries

```python
NoSQL databases (MongoDB)
3. Advanced MQL queries
We know how to do the basic operations on the MongoDB Shell, but it is possible to do much more with the MQL.

a) Operators
On MongoDB, it is possible to filter documents according to numerical data. Let's take the zips collection where we have information about the number of inhabitants of a city. It will be possible to retrieve the documents listing the documents where the population is above a specific threshold.



Operator	Description
$eq	Equal
$ne	Not equal
$gt	Greater than
$lt	Less than
$gte	Greater than or equal to
$lte	Less than or equal to
Let's see how they work with the example below. We want to retrieve all the documents where we have a population greater than or equal to 12678.

db.zips.find({"pop" : {$gte : 12678}})
There are also conditional operators to put multiple conditions on our queries. For example, we might want to display cities with a population between 10000 and 20000. To do this, we will have to select the cities with a population greater than 10000 and less than 20000.


Operator	Description
$and	AND logical
$or	OR logical
$nor	NOR logical
$not	NOT logical
Let's filter the documents according to the interval given as an example, for this we will need the $and operator and we use it in the following way:

db.zips.find({
  $and : [
    {"pop":{$lt : 20000}}, 
    {"pop":{$gt : 10000}}]
    }
  )
To do the logical AND, it is possible to do it implicitly. That is, without the $and operator, let's take the query with the interval [10000,20000], without the $and operator, we would have written :

db.zips.find(
  {
    "pop": { $lt: 20000 },
    "pop": { $gt: 10000 }
  }
)
It is also possible to compare the values of the keys of the same document with the help of the keyword $expr, let us see its use by detecting the documents of the collection grades where the values of the keys class_id and student_id are equal.

db.grades.find(
  {
    "$expr": {$eq : ["$student_id","$class_id"]}
  }
)
b) The "Cursor" methods
The result of the find function is a list of documents, but it corresponds to what we call a Cursor. Here are some useful functions on this object. We can find the list of available functions here.

  The name Cursor comes from the fact that we are not working directly on the collection but on the result of a query that we call Cursor.
sort
As its name suggests, the sort function is used to sort the documents returned by the find function. We simply need to specify on which key we want to sort our documents. This key can be numeric or a string. Let's see an example:

db.grades.find().sort({"student_id":-1})
Here, we sort our documents according to the key student_id and we put them in descending order via the value -1. We put in ascending order by putting 1 instead of -1. It is also possible to sort our documents according to two keys, it will simply have to be added in the dictionary given as argument to the sort method.

db.grades.find().sort({"student_id":-1,"class_id":1})
limit
The find function returns up to the first 20 documents, sometimes we want to retrieve a fragment in the same way as the head function of pandas. We can do this with the limit function:

db.zips.find().limit(5)
Here we return only 5 documents. We can also do this after the sort function:

db.zips.find().limit(5).sort({"city":1})
  The order between sort and limit does not matter, we will have the same documents in both cases.
count
Finally, it is possible to count the number of documents returned by the query using the count function:

db.zips.find().count()
c) Projection
From the beginning, we retrieve the whole document but we may be interested in only one key of the document. In the same way as with relational databases, it is possible to perform projections. We will see how to do this with the following query. In the zips collection, let's display the populations of the cities in the state of Texas with the command below:

db.zips.find({"state":"TX"},{"pop":1})
The first part will be used to filter the documents and the second part will be used to filter the keys we are interested in. We can make a parallel with the function .loc of the pandas library of Python. The result of the query returns only the populations of the cities in Texas with the document id.

  If you do not want to filter on the lines, it is possible to put empty braces {} in the first part of the function find.
db.zips.find({},{"pop":1})
By associating the value 1 to the key pop, we tell MongoDB to include only the number of inhabitants in our results. It is possible to do the opposite operation and exclude the key pop from our results, we will have to associate the value 0 to pop. Here is an overview of the command db.zips.find({"state": "TX"},{"pop":0}) :

[
  {
    _id: ObjectId("5c8eccc1caa187d17ca74cf7"),
    city: 'CARROLLTON',
    zip: '75006',
    loc: { y: 32.965736, x: 96.882464 },
    state: 'TX'
  },
  {
    _id: ObjectId("5c8eccc1caa187d17ca74cfb"),
    city: 'PLANO',
    zip: '75024',
    loc: { y: 33.075211, x: 96.784307 },
    state: 'TX'
  },
The documents do not contain the key pop. To collect several keys, we will have to add the key wished following the others:

Display the names and populations of cities in the state of Texas :

db.zips.find({"state":"TX"},{"pop":1,"city":1})
  Be careful, it is not possible to put 0's and 1's in the part dedicated to the projection. The only exception is when we want to remove the index from our results. We can write {_id:0}
We frequently have an array or a dictionary as value of a key. It is possible to retrieve its "sub-elements" using dot notation. Let's illustrate this with a new collection called cie giving information about companies.

Import the data to the cie collection of the sample database from the companies.json file:

mongoimport -d sample -c cie --file /data/db/companies.json
Display a document from the cie collection:

use cie

db.cie.findOne()
As we can see, we have a rather long document but with tables and sub-dictionaries. Let's look at the document where the key name is Sony and the key acquisitions, we have the following result:

sample> db.cie.find({ "name": "Sony" }, { "acquisitions": 1 })
[
  {
    _id: ObjectId("52cdef7c4bab8bd675297e6f"),
    acquisitions: [
      {
        price_amount: 260000000,
        price_currency_code: 'USD',
        term_code: null,
        source_url: 'http://www.techcrunch.com/2008/04/22/sony-buys-gracenote-for-260m/',
        source_description: 'Sony Buys Gracenote for $260M',
        acquired_year: 2008,
        acquired_month: 4,
        acquired_day: 22,
        company: { name: 'Gracenote', permalink: 'gracenote' }
      },
      {
        price_amount: null,
        price_currency_code: 'USD',
        term_code: null,
        source_url: 'http://www.freshnews.com/news/273657/sony-electronics-acquires-convergent-media-systems',
        source_description: 'Sony Electronics Acquires Convergent Media Systems',
        acquired_year: 2010,
        acquired_month: 1,
        acquired_day: 28,
        company: {
          name: 'Convergent Media Systems',
          permalink: 'convergent-media-systems'
        }
      },
      {
        price_amount: null,
        price_currency_code: 'USD',
        term_code: null,
        source_url: 'http://paidcontent.org/article/419-sony-buys-developer-of-hit-game-littlebigplanet/',
        source_description: 'Sony Buys Developer Of Hit Game LittleBigPlanet',
        acquired_year: 2010,
        acquired_month: 3,
        acquired_day: 2,
        company: { name: 'LittleBigPlanet', permalink: 'littlebigplanet' }
      },
      {
        price_amount: 380000000,
        price_currency_code: 'USD',
        term_code: null,
        source_url: 'http://allthingsd.com/20120702/sony-acquires-cloud-gaming-company-gaikai-for-380-million/',
        source_description: 'Sony Acquires Cloud-Gaming Company Gaikai For $380 Million',
        acquired_year: 2012,
        acquired_month: 7,
        acquired_day: 2,
        company: { name: 'Gaikai', permalink: 'gaikai' }
      }
    ]
  }
]
The key acquisitions is an array of dictionaries. To access the different elements of our array, we go through the index as in a classical array. However the index is not between brackets, but after a dot, so acquisitions.0 will represent the first element of our array.

Since the elements are dictionaries, we have more keys to collect. If we want to know the acquisition date of the first company in the acquisitions array, we will do acquisitions.0.acquired_year

Display the names and years of creation of companies that acquired their first *"outside" company after 2007:

db.cie.find(
  { "acquisitions.0.acquired_year": {$gt : 2007} }, 
  {"name":1,"founded_year":1,"_id" : 0}
  )
Here we have worked on the first element of the key acquisitions but we may need to see in detail all the elements of the array. To iterate over the elements we can use the $elemMatch operator, let's see how it works.

For example, let's try to find out who acquired the company LittleBigPlanet. We know that it is Sony thanks to the result of the command: db.cie.find({ "name": "Sony" }, { "acquisitions": 1 }). We will use the different keys associated with the acquisition date: acquired_year, acquired_month and acquired_day.

db.cie.find(
  {"acquisitions":
    {"$elemMatch":
      {"acquired_year":2010, 
       "acquired_month": 3,
       "acquired_day":2}}},
  {"name":1} 
)
Here we have two documents that are returned where the company name is indeed Sony.

  It would have been easier to use the name of the acquired company but it is not possible with the find method because it is nested in the company key . We have to use the aggregations that we will discuss in the next chapter.
Display the names of companies that have acquired businesses with a currency other than the dollar:

db.cie.find(
  {"acquisitions": 
    { "$elemMatch": 
      {"price_currency_code" : {"$ne":"USD"}}}},
  {"name":1, "_id":0}
)
```

# Bolum 4 -  Aggregations and Indexes

```python
4. Aggregations and Indexes
a) Aggregation
We have seen the possibilities of retrieving data with MQL, but we encountered some limitations. For example, we saw on the previous chapter, the inability of accessing a key located in a nested field. To have more possibilities, we use the aggregation framework.

From the aggregation, we can set up pipelines. A pipeline consists of several steps (stage). The way it works is different from MQL where we were looking for the desired documents. Here we will work directly from the collection where we will pass the documents to our pipeline, this is closer to functional programming. Let's look briefly at a first example of aggregation:

db.cie.aggregate([
  {$match : {"offices.city":"Seattle"}},
  {$project : {"_id":1,"name":1}}
  ]
)
Here we are retrieving documents from the cie collection where offices are located in the city of Seattle. From these documents we only want the document id and the company name.

An aggregation starts with the aggregate function that we apply to a collection and that takes an array of steps. We can put as many steps as we want, giving the possibility to make more queries as stated in the preamble.

Let's focus on the most important steps in a pipeline:

match
The first stage we'll cover is match. At first glance, it seems equivalent to the find function where we look for documents that match what we want. However, we should think of the match keyword as a filter, we run the entire collection through the filter and in the end we get back only the documents we want.

In the previous chapter, we had trouble finding the company that acquired LittleBigPlanet but with the aggregation framework we would have found Sony as follows:

db.cie.aggregate([
  {$match : {"acquisitions.company.name":"LittleBigPlanet"}}
  ]
)
We don't need to iterate over the array of the acquisitions key. Using the dot notation, we can check in the array with company key if there is if an element has the value LittleBigPlanet in the name key.

We can couple this match with the boolean operators we saw in MQL. For example, let's try to find the companies where Elon Musk was an employee. To do this, we'll use the relationships key with the first_name and last_name keys of the person key.

db.cie.aggregate([
  {$match : 
    {$and : 
      [
        {"relationships.person.first_name":"Elon"}, 
        {"relationships.person.last_name":"Musk"}
      ]
    }
  }]
)
We have documents, but it is difficult to check if they are the right documents. For this we use the project operator.

project
Let's retrieve only the company names from the previous query:

db.cie.aggregate([
  {$match : 
    {$and : 
      [
        {"relationships.person.first_name":"Elon"}, 
        {"relationships.person.last_name":"Musk"}
      ]
    }
  }, 
  {$project : {"_id":1,"name":1}}
  ]
)
We have 4 results including the document ID and the company name. We can think that this is a simple projection like with the find function but it is not the case.

We will use it to format our results. Indeed, from the returned document, instead of having name, we could write society :

db.cie.aggregate([
  {$match : 
    {$and : 
      [
        {"relationships.person.first_name":"Elon"}, 
        {"relationships.person.last_name":"Musk"}
      ]
    }
  }, 
  {$project : {"_id":1,"society":"$name"}}
  ]
)
Here we assign the value of the key name to the new key society. It is also possible to assign a dictionary to a new key:

db.cie.aggregate([
  {$match : 
    {$and : 
      [
        {"relationships.person.first_name":"Elon"}, 
        {"relationships.person.last_name":"Musk"}
      ]
    }
  }, 
  {$project : 
    {"_id":1,"society":"$name",
    "employee":
      {"first_name":"$relationships.person.first_name",
      "last_name":"$relationships.person.last_name"
    }}
  }
  ]
)
Modify the previous query, in the returned results, the boss key should contained a leader key that indicates the name of the director (we assume here that it is the first element of relationships), so you can use the title key ). You will need the functions concat and arrayElemAt.

db.cie.aggregate([
  {$match : 
    {$and : 
      [
        {"relationships.person.first_name":"Elon"}, 
        {"relationships.person.last_name":"Musk"}
      ]
    }
  }, 
  {$project : 
    {"_id":1,"society":"$name",
    "employee":
      {"name": {$concat : [
        {$arrayElemAt : ["$relationships.person.first_name",0]}," ",
        {$arrayElemAt : ["$relationships.person.last_name",0]}]},
      "title":{$arrayElemAt : ["$relationships.title",0]}    }
  }}
  ]
)
group
The last stage we'll look at is group and as the name suggests, we're going to make up packages of documents like the GROUP BY command in SQL. Let's group the companies according to their year of foundation:

db.cie.aggregate(
  [
    {
      $group: {"_id":"$founded_year"}
    }
  ]
)
It is possible to sort with the $sort operator which is different from the sort function:

db.cie.aggregate(
  [
    {
      $group: {"_id":"$founded_year", "nb_cie":{$sum:1}}
    },
    {
      $sort: {nb_cie:-1}
    }
  ]
)
Here, -1 means that we sort in descending order, while 1 allows us to sort in ascending order.

Combine the operators match, group and sort to get the number of companies whose founding year is greater than 1999 and which have premises in Paris.

db.cie.aggregate(
  [ {
      $match: {"offices.city":"Paris","founded_year":{$gt:1999}}
    },
    {
      $group: {"_id":"$founded_year", "nb_cie":{$sum:1}}
    },
    {
      $sort: {nb_cie:-1}
    }
  ]
)
b) Indexes
An index is used to improve the performance of our queries. Indeed, let's imagine we are in the library and we are looking for "The Promise of the Dawn " by Romain Gary. We will go to the letter G and then search for Gary and finally "The Promise of Dawn ". Our search is more efficient this way, if we didn't have this index, we would have taken much longer to retrieve the book.

That's why we're going to add an index to our collections. To do this, we will choose a key of the document as our index. This attribute should be chosen according to the queries we perform.

Let's illustrate with the zips collection, let's assume that we make our queries according to the name of the cities. To create our index, we will use the following line:

db.zips.createIndex({"city":1})
Before that, let's observe the performance on the query below with the function explain("executionStats") :

db.zips.find({"city":"BOSTON"},{"city":1}).explain("executionStats")
Let's look at the executionStats section:

executionStats: {
    executionSuccess: true,
    nReturned: 17,
    executionTimeMillis: 8,
    totalKeysExamined: 0,
    totalDocsExamined: 29470,
    executionStages: {
      stage: 'PROJECTION_SIMPLE',
      nReturned: 17,
      executionTimeMillisEstimate: 0,
      works: 29472,
      advanced: 17,
      needTime: 29454,
      needYield: 0,
      saveState: 29,
      restoreState: 29,
      isEOF: 1
The query took 8 ms, let's look at the performance after index:

executionStats: {
    executionSuccess: true,
    nReturned: 17,
    executionTimeMillis: 0,
    totalKeysExamined: 17,
    totalDocsExamined: 17,
    executionStages: {
      stage: 'PROJECTION_SIMPLE',
      nReturned: 17,
      executionTimeMillisEstimate: 0,
      works: 18,
      advanced: 17,
      needTime: 0,
      needYield: 0,
      saveState: 0,
      restoreState: 0,
      isEOF: 1
We are faster because we use an index.

In our previous scenario where we were looking for the book by Romain Gary, we went to the G section of the library. To optimize our search, we could first go by the category of our book, and then search by the author's name. We can do the same on MongoDB with the Compound Index.

We can illustrate this again with the zips collection by filtering by state, then filtering on city names. We create this index composed of two keys in the following way:

db.zips.createIndex({"state":1,"city":1})
Thus, by specifying the name of the state, the query is more efficient.
```

# Bolum 6. Data modeling with MongoDB

```python
6. Data modeling with MongoDB
a) Case study

So far, we have mainly handled the data. Either with MQL, the aggregation framework or via Python, however we are not interested in the structure of our documents.

We imported directly from the JSON files, but depending on the structure of the documents, the performance of the queries may be reduced.

This is why it is very important to question the modeling of data, just as with relational databases. We must also remember that data usage can influence our choices in terms of data structure. Indeed, the normalized data of relational databases are adapted for CRUD operations, but to do analyses it is common to change the structure of the data, notably by using a star schema. This star schema depends on the usage and therefore there are several ways to denormalize the data.

On MongoDB, there are no normalization rules but there is a certain order to respect to structure our documents:

    Create the design of the application

MongoDB is used to answer a query coming from an application (website, IoT sensor registration,...) to fulfil a particular need. The usage of our data is therefore defined by the data that our application requires and the read and write rules around this data.

    Define the data model We can now structure our documents according to the usage determined earlier.

    Improve the application There are always new features to add to an application, so we modify the usage defined in the first step.

    Improve the data model

These last two steps are complementary and iterative. We find again the philosophy of the pipeline CI-CD which limits the risk when we update our application.
b) Methodology

The steps described below allow to implement what has been described in the previous section.

    Evaluate the application workload.

During this step, it will be imperative to identify the business needs. This is when the business experts will be involved by making "predictions on their data".

At the end of this step, we will be able to determine the size of our data and list the operations in order of importance.

    Define entities and relationships

Here, this step is similar to what we would have done with relational databases by making UML diagrams to model our data. In particular, we will see how the collections, which represent our entities, are linked together, these will be our relationships.

For example, let's imagine that we have information about music, we can represent the data of a song in the following way:

{
  '_id': 1,
  'name' : 'Gimme Gimme Gimme',
  'singer' : 'ABBA', 
  'nationality' : 'Swedish',
  'lang' : 'English'
}
{
  '_id': 2,
  'name' : 'Dancing Queen',
  'singer' : 'ABBA', 
  'nationality' : 'Swedish',
  'lang' : 'English'
}
{
  '_id': 3,
  'name' : 'Money, Money, Money',
  'singer' : 'ABBA', 
  'nationality' : 'Swedish',
  'lang' : 'English'
}

We put all the information on the same document but there is some repeating data. We may therefore have two collections, one only on the music and one on the musicians.

{
  _id : 1
  'name' : `ABBA`
  'song_id' : [1,2,3],
  'nationality' : 'Swedish',
  'lang' : 'English'
}

{
  '_id' : 1
  'name' : 'Gimme Gimme Gimme'
  'singer_id' : 1
}
{
  '_id' : 2
  'name' : 'Dancing Queen'
  'singer_id' : 1
}
{
  '_id' : 3
  'name' : 'Money, Money, Money'
  'singer_id' : 1
}

We are once again faced with the relationships 1-1 (One to One), 1-N (One to Many) and N-N (Many to Many).

    Finalize the data model for each collection.

Once the identification of the collections and relationships is done, we can ask ourselves about the structure of the documents. Is it the most suitable for our usage?

By the structure of our document, we mean the value of our keys. For example, in some cases we had data inserted into a table, and sometimes data was inserted directly into the document.

At the end of this process, the data structure is adapted to the use of the data.
```

```python
# operators 

Operator	Description
$eq	Equal
$ne	Not equal
$gt	Greater than
$lt	Less than
$gte	Greater than or equal to
$lte	Less than or equal to


Operator	Description
$and	AND logical
$or	OR logical
$nor	NOR logical
$not	NOT logical


## Filter - 

# 01.
db.zips.find({"pop" : {$gte : 12678}})

# 02.
db.zips.find({
  $and : [
    {"pop":{$lt : 20000}}, 
    {"pop":{$gt : 10000}}]
    }
  )


# $expr  = student_id ile class_id esit olanlari buluyoruz, Kolon karasilastirmasi olmasi icin $expr ifadesi olmak zorunda
db.grades.find( { "$expr": {$eq : ["$student_id", "$class_id"]}} 
               )

db.zips.find({ $expr : {$eq : ["$zip", "$pop"]}}
             )

# sort =  -1 DESC , 1 ASC demektir
db.grades.find().sort({"student_id" : -1})

# limit()  = 
db.zips.find().sort({"pop" : -1}).limit(5)

# count()
db.zips.find().count()

# ---------------- Kolon (alan) Secim Islemleri --------------------
# Tek kolon (alan) secmek  ; (1 : True anlaminda)
db.zips.find({},{"pop":1})

# iki kolon (alan) secmek 
db.zips.find({}, { "pop": 1, "city": 1 })

# Kolon secmemek ( 0 : False demek)
db.zips.find({"state": "TX"},{"pop":0})

# Hata veriyor !!!! 0 ve 1 ayni anda olamiyor !!! 
# ISITISNA ISITISNA ISITISNA  _id = 0 oluyor.
db.zips.find({"state":"TX"}, {"city":1, "pop":0})   # ===> Hata veririr
db.zips.find({"state":"TX"}, {"city":1, "_id":0})   # ===> Hata VERMEZ 



# ---------------- AGGREGATE  ----------------
# $match (filter) and $project (kolon secer)

db.cie.aggregate([
  {$match : {"offices.city":"Seattle"}},
  {$project : {"_id":1,"name":1}}
  ]
)

# $match (project olmayinca butun veriyi getirir)
db.cie.aggregate([
  {$match : {"acquisitions.company.name":"LittleBigPlanet"}}
  ]
)

# $match and $and
db.cie.aggregate([
  {$match : 
    {$and : 
      [
        {"relationships.person.first_name":"Elon"}, 
        {"relationships.person.last_name":"Musk"}
      ]
    }
  }]
)

# $match, $and ve $project
db.cie.aggregate([
  {$match : 
    {$and : 
      [
        {"relationships.person.first_name":"Elon"}, 
        {"relationships.person.last_name":"Musk"}
      ]
    }
  }, 
  {$project : {"_id":1,"name":1}}
  ]
)

"""
{ "_id" : ObjectId("52cdef7c4bab8bd675297e89"), "name" : "PayPal" }
{ "_id" : ObjectId("52cdef7c4bab8bd675298198"), "name" : "Tesla Motors" }
{ "_id" : ObjectId("52cdef7d4bab8bd675299fe8"), "name" : "zip2" }
{ "_id" : ObjectId("52cdef7d4bab8bd675299fe9"), "name" : "SolarCity" }
"""

# $match, $and, $project ve $KOLON_ADI 
# (kolon adinin yeniden adlandirir ama sadece cikti etkilenir orjinal kolon adi ayni kalir)
db.cie.aggregate([
  {$match : 
    {$and : 
      [
        {"relationships.person.first_name":"Elon"}, 
        {"relationships.person.last_name":"Musk"}
      ]
    }
  }, 
  {$project : {"_id":1,"society":"$name"}}
  ]
)
# OUTPUT 
"""
{ "_id" : ObjectId("52cdef7c4bab8bd675297e89"), "society" : "PayPal" }
{ "_id" : ObjectId("52cdef7c4bab8bd675298198"), "society" : "Tesla Motors" }
{ "_id" : ObjectId("52cdef7d4bab8bd675299fe8"), "society" : "zip2" }
{ "_id" : ObjectId("52cdef7d4bab8bd675299fe9"), "society" : "SolarCity" }
"""


## $match, $and, $project ve $KOLON_ADI 
db.cie.aggregate([
  {$match : 
    {$and : 
      [
        {"relationships.person.first_name":"Elon"}, 
        {"relationships.person.last_name":"Musk"}
      ]
    }
  }, 
  {$project : 
    {"_id":1,"society":"$name",
    "employee":
      {"first_name":"$relationships.person.first_name",
      "last_name":"$relationships.person.last_name"
    }}
  }
  ]
)

#OUTPUT
"""
"_id" : ObjectId("52cdef7c4bab8bd675297e89"),
	"society" : "PayPal",
	"employee" : {
		"first_name" : [
			"Sam",
			"Mary",
			"John",
            ..........],
		"last_name" : [
			"Hamilton",
			"Hentges",
			"Muller",
			"Maller",
            .........
            		]
	}
}
{
	"_id" : ObjectId("52cdef7c4bab8bd675298198"),
	"society" : "Tesla Motors",
	"employee" : {
		"first_name" : [
			"Elon",
			"Marc",
			"Jerome",
			"Jay",
            ........     """ # Paypal, Tesla Motors, zip2, SolarCity firmalasri icin id,society, first_name, last_name alan adlari ciktisini verir
```

```python
# Regex
{"relationships.person.first_name": {$regex : /^elo/}}

# icinde lon gecen
{$regex : /lon/}

# Abdu ile baslayan
{$regex : /^Abdu/}

# Abdu ile baslayan  case not sensitive (incase sensitive)
{$regex : /^Abdu/i}

# {"relationships.person.first_name":/^P/}   ====>   P ile baslayan isimler
{$regex : /^P/}


# $in (elo ile baslayayack yada icinde lon olacak)
{$in : [/^elo/i, /lon/i]}


# ORNEK
db.cie.aggregate([
  {$match : 
    {$and : 
      [
        {"relationships.person.first_name": {$in : [/^elo/i, /lon/i]}}, 
        {"relationships.person.last_name":"Musk"},   
      ]
    }
  }, 
  {$project : 
    {"_id":0,"society":"$name"
    }}
  
]).pretty()



# Ornek
db.cie.aggregate([
  {$match :  {"name" : {$regex : /A/i}} },
  {$group : {"_id": "$founded_year"} },
  {$sort : {"founded_year":-1}}
]
)
```

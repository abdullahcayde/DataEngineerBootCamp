# Recap Webscraping

```python
url = 'https://www.amazon.com.au/gp/bestsellers/books/ref=zg_bs_nav_0'
page = urlopen(url)

soup = bs(page, "html.parser")

list_book = soup.findAll('div', {'class' : 'zg-grid-general-faceout'})

list_title, list_writer, list_price, list_star, list_link = [], [], [], [], []
list_book = soup.findAll('div', {'class' : 'zg-grid-general-faceout'})

for i in list_book:
    title = i.find('img')['alt']
    writer = i.find('div', {'class' : 'a-row a-size-small'}).text
    price = i.find('span', {'class': "p13n-sc-price"}).text
    try:
        star = i.find('div', {'class' : 'a-icon-row'}).text
    except:
        star = np.nan
    link = i.find('a')['href']
    list_title.append(title)
    list_writer.append(writer)
    list_price.append(price)
    list_star.append(str(star)[:3])
    list_link.append(f'https://www.amazon.com.au/{link}')

# Create DataFrame 01
dict = {'title': list_title, 'writer':list_writer, 'price':list_price, 'star':list_star, 'link':list_link}
df = pd.DataFrame(dict)  

# Create DataFrame 02
df = pd.DataFrame(list(zip(list_title, list_writer, list_price, list_star, list_link)),
                  columns=["title", "writer", "price", "star", "link"])


print(df.shape)
df.head(10)
```

```python
url_imdb = 'https://assets-datascientest.s3.eu-west-1.amazonaws.com/IMDB_en.html'

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:109.0) Gecko/20100101 Firefox/112.0'}

res = requests.get(url_imdb, headers=headers).text

soup_idm = BeautifulSoup(res, 'html.parser')

list_title = soup_idm.findAll('td', {'class' : 'titleColumn'})
list_rank = soup_idm.findAll('td', {'class': 'ratingColumn imdbRating'})

titles, years, ranks = [], [], []
for i,r in zip(list_title[:], list_rank[:]):
    title = i.text[:].split('\n')[1]
    year = i.text[:].split('\n')[2].replace('(', '').replace(')', '')
    rank = r.text.split()
    titles.append(title)
    years.append(year)
    ranks.append(rank)
    
df_imdb = pd.DataFrame(list(zip(titles, years, ranks)), columns = ['title', 'year', 'rating'])


for i in range(100):
    if df_imdb['rating'][i] != []:
        df_imdb['rating'][i] = df_imdb['rating'][i][0]
    elif df_imdb['rating'][i] == []:
        df_imdb['rating'][i] = np.nan


df_imdb['rating'] = df_imdb['rating'].str.replace(',', '.').astype('float', errors='ignore').apply(lambda x : x * 10)
df_imdb.head(15)
```

```python
from urllib.request import urlopen 
from bs4 import BeautifulSoup as bs 
import pandas as pd
import numpy as np
# Make a request to the webpage
url = "https://assets-datascientest.s3.eu-west-1.amazonaws.com/IMDB_en.html"
page = urlopen(url)

# Create a BeautifulSoup object
soup = bs(page.content, "html.parser")

# Find the section of the page with the movie information
movie_section = soup.find("div", class_="lister-list")

# Extract the movie data and store it in a list of dictionaries
movies = []
for movie in movie_section.find_all("div", class_="lister-item"):
    title = movie.find("a").get_text()
    rank = movie.find("span", class_="lister-item-index").get_text()
    rating = movie.find("span", class_="ipl-rating-star__rating").get_text()
    year = movie.find("span", class_="lister-item-year").get_text().replace("(", "").replace(")", "")
    movies.append({"title": title, "rank": rank, "rating": rating, "year": year})

# Convert the list of dictionaries into a pandas dataframe
movies_df = pd.DataFrame(movies)

# Print the dataframe
print(movies_df)
```

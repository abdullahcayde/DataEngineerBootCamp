```python


Exam 111

DataViz with Matplotlib

    The exercise is made of several questions, please answer them in order and pay attention to the variable names.

    The purpose of this exercise is to implement some visualization techniques using Matplotlib. The packages used will be pandas and Matplotlib. This exercise is made of two independent parts and tests your abilities on the concepts learned in the corresponding module.

    To help you, do not hesitate to use the Matplotlib Cheat Sheet created by us.

1. Problem presentation : Real estate exploration

    In this study, we will focus on a dataset of houses with their associated characteristics (sale price, living area, neighborhood life, garage presence, roof materials, etc). Using Matplotlib, we will visualize and study the distribution and correlations between these different data.

2. Loading the data

    (a) Load the pandas package as pd and the matplotlib.pyplot sub-package as plt.
    (b) Read the house_pricing.csv file into a DataFrame called df.
    (c) Display the first 5 rows of the df DataFrame.

 %matplotlib inline

import matplotlib.pyplot as plt

import numpy as np

import pandas as pd

df = pd.read_csv('house_pricing.csv')

​

df.head()

	Id 	MSSubClass 	MSZoning 	LotFrontage 	LotArea 	Street 	Alley 	LotShape 	LandContour 	Utilities 	... 	PoolArea 	PoolQC 	Fence 	MiscFeature 	MiscVal 	MoSold 	YrSold 	SaleType 	SaleCondition 	SalePrice
0 	1 	60 	RL 	65.0 	8450 	Pave 	NaN 	Reg 	Lvl 	AllPub 	... 	0 	NaN 	NaN 	NaN 	0 	2 	2008 	WD 	Normal 	208500
1 	2 	20 	RL 	80.0 	9600 	Pave 	NaN 	Reg 	Lvl 	AllPub 	... 	0 	NaN 	NaN 	NaN 	0 	5 	2007 	WD 	Normal 	181500
2 	3 	60 	RL 	68.0 	11250 	Pave 	NaN 	IR1 	Lvl 	AllPub 	... 	0 	NaN 	NaN 	NaN 	0 	9 	2008 	WD 	Normal 	223500
3 	4 	70 	RL 	60.0 	9550 	Pave 	NaN 	IR1 	Lvl 	AllPub 	... 	0 	NaN 	NaN 	NaN 	0 	2 	2006 	WD 	Abnorml 	140000
4 	5 	60 	RL 	84.0 	14260 	Pave 	NaN 	IR1 	Lvl 	AllPub 	... 	0 	NaN 	NaN 	NaN 	0 	12 	2008 	WD 	Normal 	250000

5 rows × 81 columns

    (d) What is the percentage of missing values in the dataset ?

df.isna().mean()*100

Id                0.000000
MSSubClass        0.000000
MSZoning          0.000000
LotFrontage      17.739726
LotArea           0.000000
                   ...    
MoSold            0.000000
YrSold            0.000000
SaleType          0.000000
SaleCondition     0.000000
SalePrice         0.000000
Length: 81, dtype: float64

( df.isna().sum().sum())

6965

3. Histograms
3.1 Distribution Visualization

    The column SalePrice represents the sale price.

    (a) Using a histogram, analyze the distribution of prices on the market. We will separate the values taken by SalePrice into 30 consecutive distinct intervals of the same length.
    (b) Use a pleasant and easy-to-understand style. Don't forget to add a legend and a title to your graphs!

#Insert code here

​

df.hist(column=['SalePrice'], bins = 30, rwidth = 0.8 , color= ['r'], label = 'Price', figsize=(10,7))

plt.title('Sale Prices Distribution')

plt.legend();

3.2 Correlations and histograms

    We will focus on the distribution of GrLivArea, which represents the living area. This distribution seems very similar to that of SalePrice. We will therefore compare these distributions between the living area and the prices of the houses in the studied dataset.

    In the following questions, we will try to superimpose these two distributions and conclude on the advantages and limitations of histograms in the study of inter-variable correlation.

    Warning: the ranges of values taken by the variables SalePrice and GrLivArea are different. It is therefore necessary to normalize those Series in order to compare their distribution and draw relevant correlation conclusions.

    Note: Normalization is a very common practice in Data Science and Data Analysis. It consists of changing the data range of variables so that variables can be compared to each other. For example, the selling prices of a housing and the living areas have two distinct ranges of values (from 10,000 to several million for the selling price of a housing in euros and from 5 to 500 for the living area in m²). The most common Machine Learning algorithms assume normalized data as input. You will see this data preprocessing step in more detail in the following modules!

    (c) Execute the following cell to normalize the data of SalePrice and GrLivArea according to Min Max normalization.

#Normalisation Min Max

df['SalePrice'] = (df['SalePrice']-df['SalePrice'].min())/(df['SalePrice'].max() - df['SalePrice'].min())

df['GrLivArea'] = (df['GrLivArea']-df['GrLivArea'].min())/(df['GrLivArea'].max() - df['GrLivArea'].min())

    Now that both variables have the same data range, we will be able to overlay their distributions.

    (d) On the same plot, display the distribution histograms of the two variables SalePrice and GrLivArea. Make sure to label your plot appropriately.

plt.figure(figsize=(10,8))

plt.hist([df.GrLivArea, df.SalePrice], bins = 7, color = ['#f27750', '#f7bf59'],

         label = ['Living Area ', 'Sale Prices'])

​

plt.title('Living Area and Sale Prices')

plt.legend();

    (e) Analyze the above figure.

# There is a positive correlation between area and price. 

​

​

​

3.3 2D Histogram

    It is also possible to create two-dimensional histograms with the hist2d function from the matplotlib.pyplot package. The principle is as follows : the abscissa and ordinate represent the two variables to be tested and the color of the box represents the number of occurrences.

    The hist2d function takes as arguments the two variables to be studied and has the same customization arguments as the hist function.

    (a) Create a 2D histogram of the variables SalePrice and GrLivArea.

    Tips:

        Use the argument cmap='Blues' within the hist2D function to have a more readable color code.
        Use the function plt.colorbar() in your code to visualize the legend of the color code used.
        Play with the number of intervals separating the ranges of values of the variables considered with the bins argument to improve the readability of the graph.

# Insert code here

​

plt.hist2d(df['GrLivArea'], df['SalePrice'], bins=10, cmap = 'Blues');

​

plt.xlabel('GrLivArea')

plt.ylabel('SalePrice')

plt.title('2D Histogram of SalePrice vs GrLivArea')

​

plt.colorbar();

    (b) Interpret the graph above.

# Enter your answer below

​

​

	GrLivArea 	SalePrice
0 	1710 	208500
1 	1262 	181500
2 	1786 	223500
3 	1717 	140000
4 	2198 	250000

df[['GrLivArea', 'SalePrice']].corr()

	GrLivArea 	SalePrice
GrLivArea 	1.000000 	0.708624
SalePrice 	0.708624 	1.000000

    Histograms are very suitable for visualizing distributions, but are not very relevant for visualizing correlations between variables as we have just seen. For correlations, scatter plots are much more appropriate.

    To know for sure which graph to choose in which situation, don't hesitate to consult our dedicated Cheat Sheet !

    (c) Run the code cell below to return to the original dataset.

df=pd.read_csv("house_pricing.csv")

4. Scatter and correlation

    The TotalBsmtSF and GrLivArea columns represent respectively cellar area and living area.

        (a) Display on the same line in two different scatter plots, the sales price as a function of each of these two variables with the names of the labels.

#Insert code here

​

​

​

    (b) What is your analysis regarding these two graphs ? What appears to be the relationship between the data ?

#Enter your answer below

​

​

​

5. Boxplot and conditional distribution

    Now we are interested in the distribution of the sales price of houses according to the level of finish quality of the apartment.

    OverallQual represents the finish quality of the considered houses. It takes discrete values between 1 and 10.

    In graphical representations of statistical data, the box plot is a quick way to describe the behavior of data in the middle and at the ends of the distributions. Values outside the box plot are considered outliers.

    (a) On the same graph, display the box plots of selling prices for each quality level. For good visibility, feel free to initialize a large figure with the figsize attribute of the plt.figure() method.

#Insert code here

​

​

​

    (b) How would you explain the increasing disparity in data as a function of finish quality ?

#Enter your answer below

​

​

​

6. Synthesis

    Finally, let's synthesize the previous studies into a single graph.

    (a) Reuse the graph of price versus living area. Color the points according to the quality of finish of the house. Vary the size of the points according to the area of the basement. We will use a coefficient of opacity of 0.4.
    (b) Add a box on the graph in which you explain which variable is represented by the size of the samples and which variable is represented by their color.

    Tips:

        You can use the following figure as inspiration for your graph (the rainbow cmap was used here).
        You can display the colorbar of the graph with the code: plt.colorbar().

# Insert code here

​

​

​

    (c) Interpret the graph above.

# Enter your answer below

​

​

​

```
